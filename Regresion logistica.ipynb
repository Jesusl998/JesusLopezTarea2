```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
framingham = pd.read_csv("framingham.csv")

```

```python
print(framingham.head())
```

```python

framingham[['BMI','diabetes']].head()
```

```python
framingham[['BMI','diabetes']].plot.scatter(x='BMI',y='diabetes')
```

```python
# pruebas de parametro
w = 0.12
b = -3.6
```

```python

# puntos de la recta
x = np.linspace(0,framingham['BMI'].max(),100)
y = 1/(1+np.exp(-(w*x+b)))
```

```python
# grafica de la recta
framingham.plot.scatter(x='BMI',y='diabetes')
plt.plot(x, y, '-r')
plt.ylim(0,framingham['diabetes'].max()*1.1)
# plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("data.csv")
```

```python
print(data.head())
```

```python
X = data["precio"].values.reshape(-1, 1)
y = data["metro"].values
```

```python
modelo = LinearRegression()
```

```python
modelo.fit(X, y)
```

```python
coeficiente_r = modelo.score(X, y)
pendiente = modelo.coef_
intercepto = modelo.intercept_
```

```python
plt.scatter(X, y, label="Datos")
plt.plot(X, modelo.predict(X), color="yellow", label="Línea de regresión")
plt.xlabel("precio")
plt.ylabel("metro")
plt.legend()
plt.show()
```
